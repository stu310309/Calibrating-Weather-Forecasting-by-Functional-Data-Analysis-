```python
def last_time_interval(d):
    time_index_last_days = np.empty(time_index.shape)
    time_index_last_days[:, 1] = time_index[:, 0]
    time_delta = datetime.timedelta(d)
    for i in range(time_index_last_days.shape[0]):
        first_day = datetime.date(int(time_index[i, 0]/10000), 
                                  int((time_index[i, 0]-int(time_index[i, 0]/10000)*10000)/100),
                                  int(time_index[i, 0]-int(time_index[i, 0]/100)*100)) - time_delta
        time_index_last_days[i, 0] = first_day.year*10000 + first_day.month*100 + first_day.day
    return(time_index_last_days)

def Cut_obs_data(time_index):
    t_obs = []
    temp_obs = []
    for j in range(time_index.shape[0]):
        t_j = np.array([])
        temp_j = np.array([])
        for i in range(len(data_cwb)):
            data_cwb_time = data_cwb[i][:, 0]*10000 + data_cwb[i][:, 1]*100 + data_cwb[i][:, 2] + data_cwb[i][:, 3]/24
            data_epa_time = data_epa[i][:, 0]*10000 + data_epa[i][:, 1]*100 + data_epa[i][:, 2] + data_epa[i][:, 3]/24
            if any(((data_cwb_time>=time_index[j, 0])&(data_cwb_time<time_index[j, 1]))) | any((data_epa_time>=time_index[j, 0])&(data_epa_time<time_index[j, 1])):
                cwb_index = np.array(np.where((data_cwb_time>=time_index[j, 0])&(data_cwb_time<time_index[j, 1]))).reshape(-1)
                epa_index = np.array(np.where((data_epa_time>=time_index[j, 0])&(data_epa_time<time_index[j, 1]))).reshape(-1)
                time_obs = np.vstack((data_cwb_time[cwb_index].reshape(-1, 1), data_epa_time[epa_index].reshape(-1, 1)))
                time_min = datetime.date(int(time_index[j, 0]/10000), 
                                         int((time_index[j, 0]-int(time_index[j, 0]/10000)*10000)/100),
                                         int(time_index[j, 0]-int(time_index[j, 0]/100)*100))
                for k in range(time_obs.shape[0]):
                    time_obs[k] = (datetime.date(int(time_obs[k]/10000), 
                                                 int((time_obs[k]-int(time_obs[k]/10000)*10000)/100),
                                                 int(time_obs[k]-int(time_obs[k]/100)*100)) - time_min).days + time_obs[k]-int(time_obs[k]) + 1
                t_j = np.append(t_j, np.hstack((time_obs, np.vstack((data_cwb[i][cwb_index, :][:, np.array([6, 5])],
                                                                     data_epa[i][epa_index, :][:, np.array([6, 5])]))))).reshape(-1, 3)
                temp_j = np.append(temp_j, np.vstack((data_cwb[i][cwb_index, 7].reshape(-1, 1),
                                                      data_epa[i][epa_index, 7].reshape(-1, 1))).reshape(-1))
        t_obs.append(t_j)
        temp_obs.append(temp_j)
    return([t_obs, temp_obs])
```

```python
time_index_last_7days = last_time_interval(7)
time_index_last_14days = last_time_interval(14)
time_index_last_21days = last_time_interval(21)
```
